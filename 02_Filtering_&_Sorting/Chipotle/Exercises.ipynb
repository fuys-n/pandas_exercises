### Step 3. Assign it to a variable called chipo.

```python
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'
data = pd.read_csv(url,sep = '\t') #.read_csv 默认以逗号分隔
```

```python
data.reset_index(level=0,inplace=True) #索引转化为列
data.head()
```

```python
# print(data.shape)

print(data.info())
print(data.describe())
```

### Step 4. How many products cost more than $10.00?

```python
# data.item_price.dtype
#change type of item_price
#first method
change_type = lambda x:float(x[1:-1])
data.item_price = data.item_price.apply(change_type)

# #second method
# # clean the item_price column and transform it in a float
# prices = [float(value[1 : -1]) for value in data.item_price]

# # reassign the column with the cleaned prices
# data.item_price = prices
```

```python
# print(data.item_price[data.item_price>10].count())

data_filtered = data.drop_duplicates(['item_name','quantity']) #.drop_duplicates()去重 .duplicated() 检测重复
```

```python
data_filtered_one = data_filtered[data_filtered.quantity == 1]
```

```python
data_filtered_one[data_filtered_one.item_price > 10].item_name.nunique() #.nunique()返回唯一值个数 .unique()返回唯一值
```

### Step 6. Sort by the name of the item

```python
print(data.sort_values('item_name'))
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
# data.head()
most_expensive = data.sort_values('item_price',ascending = False).head(1).quantity
print(most_expensive)
```
